## Required Libraries

```python
import pandas as pd
import json
import os
from pprint import pprint
```

## Aggregrated Transaction Data

```python
path="/Users/deepega/DS-Mainboot/Phone pulse data/pulse/data/aggregated/transaction/country/india/state/"
agg_state_list=os.listdir(path)

agg_trans_data_col = {'State':[], 'Year':[],'Quater':[],'Transacion_type':[], 'Transacion_count':[], 'Transacion_amount':[]}

for i in agg_state_list:
    p_i = path + i + "/"
    agg_yr = os.listdir(p_i)
    for j in agg_yr:
        p_j = p_i + j + "/"
        file_inside_yr = os.listdir(p_j)
        for k in file_inside_yr:
            p_k = p_j + k
            data = json.load(open(p_k,'r'))
            for z in data["data"]["transactionData"]:
                agg_trans_data_col["State"].append(i)
                agg_trans_data_col["Year"].append(j)
                agg_trans_data_col["Quater"].append(int(k.strip('.json')))
                agg_trans_data_col["Transacion_type"].append(z['name'])
                agg_trans_data_col["Transacion_count"].append(z["paymentInstruments"][0]["count"])
                agg_trans_data_col["Transacion_amount"].append(z["paymentInstruments"][0]["amount"])
                
                            
df_agg_trans_data_col = pd.DataFrame(agg_trans_data_col)
df_agg_trans_data_col
```

## Aggregated User Data

```python
path="/Users/deepega/DS-Mainboot/Phone pulse data/pulse/data/aggregated/user/country/india/state/"
agg_state_list=os.listdir(path)

agg_user_data_col = {"State":[], 'Year':[], 'Quarter':[], 'Brand': [], 'Count': [], 'Percentage':[]}
for i in agg_state_list:
    p_i = path + i + "/"
    agg_yr = os.listdir(p_i)
    for j in agg_yr:
        p_j = p_i + j + "/"
        file_inside_yr = os.listdir(p_j)
        for k in file_inside_yr:
            p_k = p_j + k
            data = json.load(open(p_k,'r'))
            try:
                for z in data["data"]["usersByDevice"]:
                    agg_user_data_col["State"].append(i)
                    agg_user_data_col["Year"].append(j)
                    agg_user_data_col["Quarter"].append(int(k.strip('.json')))
                    agg_user_data_col["Brand"].append(z["brand"])
                    agg_user_data_col["Count"].append(z["count"])
                    agg_user_data_col["Percentage"].append(z["percentage"])
                    # print(z)
            except:
                pass

df_agg_user_data_col = pd.DataFrame(agg_user_data_col)
df_agg_user_data_col
```

## Map Transaction Data

```python
path = "/Users/deepega/DS-Mainboot/Phone pulse data/pulse/data/map/transaction/hover/country/india/state/"
map_state_list= os.listdir(path)

map_trans_data_col = {"State":[], "Year":[], "Quarter":[], "District":[], "Number_of_transaction":[], "Amount":[]}
for i in map_state_list:
    p_i = path + i + "/"
    map_yr = os.listdir(p_i)
    for j in map_yr:
        p_j = p_i + j + "/"
        file_inside_yr = os.listdir(p_j)
        for k in file_inside_yr:
            p_k = p_j + k
            data = json.load(open(p_k,'r'))
            for z in data["data"]["hoverDataList"]:
                map_trans_data_col["State"].append(i)
                map_trans_data_col["Year"].append(j)
                map_trans_data_col["Quarter"].append(int(k.strip('.json')))
                map_trans_data_col["District"].append(z["name"])
                map_trans_data_col["Number_of_transaction"].append(z["metric"][0]["count"])
                map_trans_data_col["Amount"].append(z["metric"][0]["amount"])
            
df_map_trans_data_col = pd.DataFrame(map_trans_data_col)   
df_map_trans_data_col       
```

### Map User Data

```python
path = "/Users/deepega/DS-Mainboot/Phone pulse data/pulse/data/map/user/hover/country/india/state/"
map_state_list= os.listdir(path)

map_user_data_col = {"State":[], "Year":[], "Quarter":[], "District": [], "Registered_User": []}
for i in map_state_list:
    p_i = path + i + "/"
    map_yr = os.listdir(p_i)
    for j in map_yr:
        p_j = p_i + j + "/"
        file_inside_yr = os.listdir(p_j)
        for k in file_inside_yr:
            p_k = p_j + k
            data = json.load(open(p_k,'r'))
            for z in data["data"]["hoverData"].items():
                map_user_data_col["State"].append(i)
                map_user_data_col["Year"].append(j)
                map_user_data_col["Quarter"].append(int(k.strip('.json')))
                map_user_data_col["District"].append(z[0])
                map_user_data_col["Registered_User"].append(z[1]["registeredUsers"])
                
df_map_user_data_col=pd.DataFrame(map_user_data_col)
df_map_user_data_col
```

### Top Transaction Data

```python
path = "/Users/deepega/DS-Mainboot/Phone pulse data/pulse/data/top/transaction/country/india/state/"
top_state_list = os.listdir(path)

top_trans_data_col = {"State":[], "Year":[], "Quarter":[], "Pincode":[], "Transaction_count":[], "Transaction_amount":[]}

for i in top_state_list:
    p_i = path + i + "/"
    map_yr = os.listdir(p_i)
    for j in map_yr:
        p_j = p_i + j + "/"
        file_inside_yr = os.listdir(p_j)
        for k in file_inside_yr:
            p_k = p_j + k
            data = json.load(open(p_k,'r'))
            for z in data["data"]["pincodes"]:
                top_trans_data_col["State"].append(i)
                top_trans_data_col["Year"].append(j)
                top_trans_data_col["Quarter"].append(int(k.strip('.json')))
                top_trans_data_col["Pincode"].append(z["entityName"])
                top_trans_data_col["Transaction_count"].append(z["metric"]["count"])
                top_trans_data_col["Transaction_amount"].append(z["metric"]["amount"])
df_top_trans_data_col = pd.DataFrame(top_trans_data_col)
df_top_trans_data_col
```
